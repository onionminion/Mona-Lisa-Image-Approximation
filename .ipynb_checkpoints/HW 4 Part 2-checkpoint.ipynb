# Random Forests for Image Approximation

```python
import pandas as pd
import numpy as np
import random
from skimage import color, io, transform
from sklearn.ensemble import RandomForestRegressor
```

## Part A

```python
im = io.imread('MonaLisa.jpeg')
print(im.shape)
io.imshow(im)
```

```python
im_resized = transform.resize(im, (im.shape[0] // 4, im.shape[1] // 4))
io.imshow(im_resized)
print(im_resized.shape)
```

```python
height = im_resized.shape[0]
width = im_resized.shape[1]
```

## Part B: Preprocessing Input

```python
num_coord = 5000

coordinates = np.zeros([num_coord, 2], dtype=int)
for i in range(num_coord):
    y = random.randint(0, height - 1)
    x = random.randint(0, width - 1)
    coordinates[i] = [x, y]
coordinates
```

Mean subtraction, standardization and unit-normalization are not required because the random forest algorithm is a decision tree-based algorithm that does not depend on data scaling. Since random forests use thresholds to partition data into regions, we do not need to scale the threshold.

## Part C: Preprocessing Output

```python
im_gray = color.rgb2gray(im_resized)
io.imshow(im_gray, cmap="gray")
im_gray.shape
im_gray
```

```python
train = np.ones([height, width])
intensities = []
for x, y in coordinates:
    intensities.append(im_gray[y, x])
    train[y, x] = im_gray[y, x]
io.imshow(train, cmap='gray')
train
```

## Part D: Random Forest Output

```python
def random_forest(n_estimators, depth):
    regr = RandomForestRegressor(n_estimators=n_estimators, max_depth=depth, random_state=0)
    regr.fit(coordinates, intensities)

    output = np.zeros([height, width])
    for x in range(width):
        for y in range(height):
            output[y, x] = regr.predict([[x, y]])
    io.imshow(output, cmap='gray')
```
