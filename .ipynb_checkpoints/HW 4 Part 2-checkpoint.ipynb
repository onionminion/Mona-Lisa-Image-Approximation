# Random Forests for Image Approximation

```python
import pandas as pd
import numpy as np
import random
from skimage import color, io, transform
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
```

## Part A

```python
im = io.imread('MonaLisa.jpeg')
print(im.shape)
io.imshow(im)
```

```python
im_resized = transform.resize(im, (im.shape[0] // 4, im.shape[1] // 4))
io.imshow(im_resized)
print(im_resized.shape)
```

```python
height = im_resized.shape[0]
width = im_resized.shape[1]
```

Because the dimension for the original image is 1526 * 1024 * 3 could be computationally intensive, we chose to reduce the height and width by resizing the image. 

## Part B: Preprocessing Input

```python
num_coord = 5000

coordinates = np.zeros([num_coord, 2], dtype=int)
for i in range(num_coord):
    y = random.randint(0, height - 1)
    x = random.randint(0, width - 1)
    coordinates[i] = [x, y]
coordinates
```

Mean subtraction, standardization and unit-normalization are not required because the random forest algorithm is a decision tree-based algorithm that does not depend on data scaling. Since random forests use thresholds to partition data into regions, we do not need to scale the threshold.

## Part C: Preprocessing Output

```python
im_gray = color.rgb2gray(im_resized)
io.imshow(im_gray, cmap="gray")
im_gray.shape
im_gray
```

```python
train = np.ones([height, width])
intensities = []
for x, y in coordinates:
    intensities.append(im_gray[y, x])
    train[y, x] = im_gray[y, x]
io.imshow(train, cmap='gray')
train
```

We chose to convert the image into grayscale so that it is computationally less expensive. In addition, by converting the image into grayscale, we no longer need to worry about different colors. Moreover, since the `color.rgb2gray()` function rescales the pixel intensities to lie between 0.0 and 1.0, we did not have to do any additional rescaling. We chose not to do any other preprocessing.

## Part D: Random Forest Output

```python
rf = RandomForestRegressor(n_estimators=3, max_depth=10, random_state=210)
rf.fit(coordinates, intensities)

final_output = np.zeros([height, width])
for x in range(width):
    for y in range(height):
        final_output[y, x] = rf.predict([[x, y]])
io.imshow(final_output, cmap='gray')
```

The above image is the final image of the random forest model.

source: https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestRegressor.html

## Part E: Experimentation 

```python
def random_forest(n_estimators, depth, min_samples_split=2):
    regr = RandomForestRegressor(n_estimators=n_estimators, max_depth=depth, 
                                 min_samples_split=min_samples_split, 
                                 random_state=210)
    regr.fit(coordinates, intensities)

    output = np.zeros([height, width])
    for x in range(width):
        for y in range(height):
            output[y, x] = regr.predict([[x, y]])
    io.imshow(output, cmap='gray')
```

### (i) Experiment for a random forest containing a single decision tree, but with depths 1, 2, 3, 5, 10, and 15

```python
random_forest(1, 1)
```

```python
random_forest(1, 2)
```

```python
random_forest(1, 3)
```

```python
random_forest(1, 5)
```

```python
random_forest(1, 10)
```

```python
random_forest(1, 15)
```

As the depth of the tree increases, the more details the resulting image has. In other words, higher depth trees seem to result in a more diverse set of colors and smaller sized color patches. For example, the the resulting image for the random forest model with depth = 1 contains just two color blocks while that of depth = 15 contains various color patches. We can conclude that the higher depth a tree has, the more accurate the model is. However, a random forest tree model with an extremely large depth could also potentially be computationally intensive and overfitting.

### (ii) Experiment for a random forest of depth 7, but with number of trees equal to 1, 3, 5, 10, and 100.

```python
random_forest(1, 7)
```

```python
random_forest(3, 7)
```

```python
random_forest(5, 7)
```

```python
random_forest(10, 7)
```

```python
random_forest(100, 7)
```

The image produced by the random forest with just one tree has color patches that look distinct from their neighbors and this makes the image very sharp. On the other hand, it is difficult to see distinct color patches in the the image produced by the random forest with 100 trees. Furthermore, as the number of trees increases, the more layers of similar color patches are added, producing a smoother image. In conclusion, as the number of trees used in the random forest model increases, it produces a smoother image. However, again, a random forest tree model with too many trees could also be computationally intensive and potentially overfitting.

### (iii) KNN Regressor for K = 1

```python
neigh = KNeighborsRegressor(n_neighbors=1)
neigh.fit(coordinates, intensities)
output = np.zeros([height, width])
for x in range(width):
    for y in range(height):
        output[y, x] = neigh.predict([[x, y]])
io.imshow(output, cmap='gray')
```

source: https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsRegressor.html

The KNN regressor with k = 1 seems to produce more accurate image of Mona Lisa than the random forest regressor. The color patches in the image produced by the KNN regressor are irregular shapes and they all look different from each other while those in the images produced by the random forest regressor are all rectangles. All of these happen because for KNN regressor with k = 1, each point depends only on one neighbor point to set the target value and thus, it is guaranteed to perform well on the training dataset.

### (iv) Experiment with different pruning strategies

```python
random_forest(1, 10, 2)
```

```python
random_forest(1, 10, 5)
```

```python
random_forest(1, 10, 10)
```

```python
random_forest(1, 10, 20)
```

For the last experiment, we decided to change the `min_samples_split` value of the random forest model. `min_samples_split` indicates the minimum number of samples required to split each node. If this value is increased, the size of each color patch seems to be also increased. For example, the image produced by the random forest model with `min_samples_split` = 2 has smaller color patches than the image produced by the random forest model with `min_samples_split` = 20. Due to the increase in the size of color patches, the image with a high `min_samples_split` value seems to lack details.

```python
from sklearn import tree

decision_tree = rf.estimators_[0]
print("One of the trained decision tree inside the forest:\n")
print(tree.export_text(decision_tree, feature_names=['width', 'height']))
```

Suppose the split point rule at the root node is $r_1(x)$, where $x$ represents a coordinate on the image.

Then the one line formula for $r_1(x)$ is the following:

$$ 
r_1(x) = 
\begin{cases}
    \text{true} & \text{if } x_\text{height} \leq 202.50 \\

### (ii) Why does the resulting image look like the way it does? What shape are the patches of color, and how are they arranged?

The final output image seems to resemble the original image closely. The image has rectangular color patches. This is because random forest models are essentially decision trees with bagging and decision trees tend to output blocky, non-smooth result. The image also seems to have multiple layers of similar color patches on top of one another. This occurs because our random forest model uses three decision trees to make final prediction of the pixel intensity in hopes to smooth out the result.
